```python
import pandas as pd
import numpy as np
np.random.seed(0)
import matplotlib.pyplot as plt
df = pd.read_csv('dados-Ambos-Tudo.csv')
nomes = df.NOME.unique()
df = df[df['UsoCpu'] < 81]
df = df[df['UsoRam'] < 81]
df.sample(frac=1).reset_index(drop=True)
df["QoE"] = df["QoE"] + 3
dadosFear = df[df['Jogo'] == 'FEAR']
df = df[df['Jogo'] != 'FEAR']
#'DelayVideo', 'DelayComandos', 'JitterVideo', 'JitterComandos', 'PerdaVideo', 'PerdaComandos'
print("------------------------------Dados gerais----------------------------------")
print("Número de jogadores: "+str(len(nomes)))
print("Total de partidas: "+str(len(df)))
HorasJogadas = float(df['TempoJogo'].sum())
HorasJogadas = HorasJogadas/60.0
HorasJogadas = np.round(HorasJogadas,2)
minutosJogados = np.round(np.round(HorasJogadas - int(HorasJogadas),2)*60,0)
print("Tempo total jogado: "+str(int(HorasJogadas))+"h e "+str(int(minutosJogados))+"m")
print("\n-------------------------------Atraso de vídeo------------------------------")
DelayVideo = df.DelayVideo.unique()
DelayVideo = np.sort(DelayVideo)
for i in range(len(DelayVideo)):
    temp = df[df['DelayVideo'] == DelayVideo[i]]
    print(str(len(temp))+" partidas com "+str(DelayVideo[i])+" ms de atraso de vídeo, "+str(np.round((float(len(temp))*100.0/float(len(df))),2))+"% do total")
print("\n-------------------------------Atraso de Comandos----------------------------")
DelayComandos = df.DelayComandos.unique()
DelayComandos = np.sort(DelayComandos)
for i in range(len(DelayComandos)):
    temp = df[df['DelayComandos'] == DelayComandos[i]]
    print(str(len(temp))+" partidas com "+str(DelayComandos[i])+" ms de atraso de comandos, "+str(np.round((float(len(temp))*100.0/float(len(df))),2))+"% do total")
print("\n-------------------------------Jitter de vídeo----------------------------")
JitterVideo = df.JitterVideo.unique()
JitterVideo = np.sort(JitterVideo)
for i in range(len(JitterVideo)):
    temp = df[df['JitterVideo'] == JitterVideo[i]]
    print(str(len(temp))+" partidas com "+str(JitterVideo[i])+" ms de jitter de vídeo, "+str(np.round((float(len(temp))*100.0/float(len(df))),2))+"% do total")
print("\n-------------------------------Jitter de comandos-------------------------")
JitterComandos = df.JitterComandos.unique()
JitterComandos = np.sort(JitterComandos)
for i in range(len(JitterComandos)):
    temp = df[df['JitterComandos'] == JitterComandos[i]]
    print(str(len(temp))+" partidas com "+str(JitterComandos[i])+" ms de jitter de comandos, "+str(np.round((float(len(temp))*100.0/float(len(df))),2))+"% do total")
print("\n-------------------------------Perda de vídeo-----------------------------")
PerdaVideo = df.PerdaVideo.unique()
PerdaVideo = np.sort(PerdaVideo)
for i in range(len(PerdaVideo)):
    temp = df[df['PerdaVideo'] == PerdaVideo[i]]
    print(str(len(temp))+" partidas com "+str(PerdaVideo[i])+"% de perda de vídeo, "+str(np.round((float(len(temp))*100.0/float(len(df))),2))+"% do total")
print("\n-------------------------------Perda de comandos-----------------------------")
PerdaComandos = df.PerdaComandos.unique()
PerdaComandos = np.sort(PerdaComandos)
for i in range(len(PerdaComandos)):
    temp = df[df['PerdaComandos'] == PerdaComandos[i]]
    print(str(len(temp))+" partidas com "+str(PerdaComandos[i])+"% de perda de comandos, "+str(np.round((float(len(temp))*100.0/float(len(df))),2))+"% do total")
print("\n--------------------------Dados dos jogadores-------------------------------")
for i in range(len(nomes)):
    temp = df[df['NOME'] == nomes[i]]
    print(str(nomes[i])+" jogou "+str(len(temp))+" partidas, "+str(np.round((float(len(temp))*100.0/float(len(df))),2))+"% do total")
print("\n---------------------------Dados dos Jogos----------------------------------")
SP = df[df['Jogo'] == 'Spelunky']
HL2 = df[df['Jogo'] == 'Half-Life 2']
print(str(len(SP))+" partidas de Spelunky ("+str(np.round((float(len(SP))*100.0/float(len(df))),2))+"% do total)")
print(str(len(HL2))+" partidas de Half-Life 2 ("+str(np.round((float(len(HL2))*100.0/float(len(df))),2))+"% do total)")
print(str(len(dadosFear))+" partidas de FEAR para validação")
print("\n-------------------------QoE TOTAL------------------------------------------")
QOE0 = df[df['QoE'] == 0]
QOE1 = df[df['QoE'] == 1]
QOE2 = df[df['QoE'] == 2]
QOE3 = df[df['QoE'] == 3]
QOE4 = df[df['QoE'] == 4]
QOE5 = df[df['QoE'] == 5]
QOE6 = df[df['QoE'] == 6]
print(str(len(QOE0))+" partidas com QoE Extremamente Péssimo ("+str(np.round((float(len(QOE0))*100.0/float(len(df))),2))+"% do total)")
print(str(len(QOE1))+" partidas com QoE Péssimo ("+str(np.round((float(len(QOE1))*100.0/float(len(df))),2))+"% do total)")
print(str(len(QOE2))+" partidas com QoE Ruim ("+str(np.round((float(len(QOE2))*100.0/float(len(df))),2))+"% do total)")
print(str(len(QOE3))+" partidas com QoE Regular ("+str(np.round((float(len(QOE3))*100.0/float(len(df))),2))+"% do total)")
print(str(len(QOE4))+" partidas com QoE Bom ("+str(np.round((float(len(QOE4))*100.0/float(len(df))),2))+"% do total)")
print(str(len(QOE5))+" partidas com QoE Excelente ("+str(np.round((float(len(QOE5))*100.0/float(len(df))),2))+"% do total)")
print(str(len(QOE6))+" partidas com QoE Ideal ("+str(np.round((float(len(QOE6))*100.0/float(len(df))),2))+"% do total)")
print("\n-------------------------QoE Somente Spelunky-------------------------------")
QOE0 = SP[SP['QoE'] == 0]
QOE1 = SP[SP['QoE'] == 1]
QOE2 = SP[SP['QoE'] == 2]
QOE3 = SP[SP['QoE'] == 3]
QOE4 = SP[SP['QoE'] == 4]
QOE5 = SP[SP['QoE'] == 5]
QOE6 = SP[SP['QoE'] == 6]
print(str(len(QOE0))+" partidas com QoE Extremamente Péssimo ("+str(np.round((float(len(QOE0))*100.0/float(len(SP))),2))+"% do total)")
print(str(len(QOE1))+" partidas com QoE Péssimo ("+str(np.round((float(len(QOE1))*100.0/float(len(SP))),2))+"% do total)")
print(str(len(QOE2))+" partidas com QoE Ruim ("+str(np.round((float(len(QOE2))*100.0/float(len(SP))),2))+"% do total)")
print(str(len(QOE3))+" partidas com QoE Regular ("+str(np.round((float(len(QOE3))*100.0/float(len(SP))),2))+"% do total)")
print(str(len(QOE4))+" partidas com QoE Bom ("+str(np.round((float(len(QOE4))*100.0/float(len(SP))),2))+"% do total)")
print(str(len(QOE5))+" partidas com QoE Excelente ("+str(np.round((float(len(QOE5))*100.0/float(len(SP))),2))+"% do total)")
print(str(len(QOE6))+" partidas com QoE Ideal ("+str(np.round((float(len(QOE6))*100.0/float(len(SP))),2))+"% do total)")
print("\n-------------------------QoE Somente Half-Life 2-----------------------------")
QOE0 = HL2[HL2['QoE'] == 0]
QOE1 = HL2[HL2['QoE'] == 1]
QOE2 = HL2[HL2['QoE'] == 2]
QOE3 = HL2[HL2['QoE'] == 3]
QOE4 = HL2[HL2['QoE'] == 4]
QOE5 = HL2[HL2['QoE'] == 5]
QOE6 = HL2[HL2['QoE'] == 6]
print(str(len(QOE0))+" partidas com QoE Extremamente Péssimo ("+str(np.round((float(len(QOE0))*100.0/float(len(HL2))),2))+"% do total)")
print(str(len(QOE1))+" partidas com QoE Péssimo ("+str(np.round((float(len(QOE1))*100.0/float(len(HL2))),2))+"% do total)")
print(str(len(QOE2))+" partidas com QoE Ruim ("+str(np.round((float(len(QOE2))*100.0/float(len(HL2))),2))+"% do total)")
print(str(len(QOE3))+" partidas com QoE Regular ("+str(np.round((float(len(QOE3))*100.0/float(len(HL2))),2))+"% do total)")
print(str(len(QOE4))+" partidas com QoE Bom ("+str(np.round((float(len(QOE4))*100.0/float(len(HL2))),2))+"% do total)")
print(str(len(QOE5))+" partidas com QoE Excelente ("+str(np.round((float(len(QOE5))*100.0/float(len(HL2))),2))+"% do total)")
print(str(len(QOE6))+" partidas com QoE Ideal ("+str(np.round((float(len(QOE6))*100.0/float(len(HL2))),2))+"% do total)")
print("\n-------------------------QoE Somente FEAR------------------------------------")
QOE0 = dadosFear[dadosFear['QoE'] == 0]
QOE1 = dadosFear[dadosFear['QoE'] == 1]
QOE2 = dadosFear[dadosFear['QoE'] == 2]
QOE3 = dadosFear[dadosFear['QoE'] == 3]
QOE4 = dadosFear[dadosFear['QoE'] == 4]
QOE5 = dadosFear[dadosFear['QoE'] == 5]
QOE6 = dadosFear[dadosFear['QoE'] == 6]
print(str(len(QOE0))+" partidas com QoE Extremamente Péssimo ("+str(np.round((float(len(QOE0))*100.0/float(len(dadosFear))),2))+"% do total)")
print(str(len(QOE1))+" partidas com QoE Péssimo ("+str(np.round((float(len(QOE1))*100.0/float(len(dadosFear))),2))+"% do total)")
print(str(len(QOE2))+" partidas com QoE Ruim ("+str(np.round((float(len(QOE2))*100.0/float(len(dadosFear))),2))+"% do total)")
print(str(len(QOE3))+" partidas com QoE Regular ("+str(np.round((float(len(QOE3))*100.0/float(len(dadosFear))),2))+"% do total)")
print(str(len(QOE4))+" partidas com QoE Bom ("+str(np.round((float(len(QOE4))*100.0/float(len(dadosFear))),2))+"% do total)")
print(str(len(QOE5))+" partidas com QoE Excelente ("+str(np.round((float(len(QOE5))*100.0/float(len(dadosFear))),2))+"% do total)")
print(str(len(QOE6))+" partidas com QoE Ideal ("+str(np.round((float(len(QOE6))*100.0/float(len(dadosFear))),2))+"% do total)")



```
